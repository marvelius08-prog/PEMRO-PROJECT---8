```python
print('Descriptive statistics for numerical columns:')
display(df.describe())
```

```python
df['spud_date'] = pd.to_datetime(df['spud_date'])
df['completion_date'] = pd.to_datetime(df['completion_date'])

print("Data types after conversion:")
display(df.info())
```

```python
df['completion_days'] = (df['completion_date'] - df['spud_date']).dt.days
df['spud_year'] = df['spud_date'].dt.year
df['production_per_meter'] = df['initial_production_bpd'] / df['depth_m']

print("New features created:")
display(df.head())
```

```python
data_np = np.array(df.index)

print('Mean Index =', np.mean(data_np))

print('Max Index =', np.max(data_np))
```

```python
df['production_category'] = pd.qcut(df['initial_production_bpd'],
                                   q=3,
                                   labels=['Low', 'Moderate', 'High'])

print("Production categories created:")
display(df[['initial_production_bpd', 'production_category']].head())
```

```python
df['production_rate_per_day'] = df['cumulative_production_bbl'] / df['completion_days']

print("New 'production_rate_per_day' column created:")
display(df[['well_id', 'cumulative_production_bbl', 'completion_days', 'production_rate_per_day']].head())
```

```python
sns.histplot(df['production_rate_per_day'].replace([np.inf, -np.inf], np.nan).dropna(), kde=True)
plt.title('Distribution of Production Rate per Day'), plt.xlabel('Production Rate (bbl/day)'), plt.ylabel('Count')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
print(" Simulated Gas-Oil Ratio (GOR) Analysis ")
df['initial_gas_production_mcfd'] = np.random.normal(loc=1000, scale=300, size=len(df))
df['initial_gas_production_mcfd'] = df['initial_gas_production_mcfd'].apply(lambda x: max(10, x))
df['gor'] = df.apply(lambda row: row['initial_gas_production_mcfd'] / row['initial_production_bpd']
                     if row['initial_production_bpd'] > 0
                     else 0, axis=1)

print("Descriptive statistics for GOR:")
display(df['gor'].describe())

sns.histplot(df['gor'], kde=True)
plt.title('Distribution of Gas Oil Ratio (GOR)'), plt.xlabel('GOR (Mcf/bbl)'), plt.ylabel('Count'), plt.grid(True, linestyle='--', alpha=0.7)
plt.show()

print("Top 5 wells by GOR:")
display(df[['well_id', 'initial_production_bpd', 'initial_gas_production_mcfd', 'gor']].sort_values(by='gor', ascending=False).head())
```

```python
df['initial_production_bpd'] = df['initial_production_bpd'].astype(float)
print("Data type of 'initial_production_bpd' after conversion:")
df.info()
```

```python
df['depth_ft'] = df['depth_m']
print("DataFrame with new 'depth_ft' column:")
display(df[['depth_m', 'depth_ft']].head())
```

```python
#Times-series
yearly_production = df.groupby('spud_year')['initial_production_bpd'].sum().reset_index()

plt.figure(figsize=(12, 6))
sns.lineplot(x='spud_year', y='initial_production_bpd', data=yearly_production)
plt.title('Total Initial Production by Spud Year')
plt.xlabel('Spud Year')
plt.ylabel('Total Initial Production (bpd)')
plt.grid(True, alpha=0.7)
plt.show()
```

```python
#bar
avg_production_by_operator = df.groupby('operator')['initial_production_bpd'].mean().reset_index()

plt.figure(figsize=(12, 7))
sns.barplot(x='operator', y='initial_production_bpd', data=avg_production_by_operator.sort_values(by='initial_production_bpd', ascending=False), palette='viridis')
plt.title('Average Initial Production per Operator')
plt.xlabel('Operator')
plt.ylabel('Average Initial Production (bpd)')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
#crossplot
plt.figure(figsize=(10, 8))
sns.scatterplot(x='depth_m', y='initial_production_bpd', data=df, alpha=0.6, hue='well_type')
plt.title('Depth vs. Initial Production (Cross-plot)')
plt.xlabel('Depth (meters)')
plt.ylabel('Initial Production (bpd)')
plt.grid(True, alpha=0.7)
plt.show()
```

```python
#boxplot
plt.figure(figsize=(10, 7))
sns.boxplot(x='production_category', y='initial_production_bpd', data=df, palette='pastel')
plt.title('Distribution of Initial Production by Production Category')
plt.xlabel('Production Category'), plt.ylabel('Initial Production (bpd)')
plt.grid(axis='y', alpha=0.7)
plt.show()
```
